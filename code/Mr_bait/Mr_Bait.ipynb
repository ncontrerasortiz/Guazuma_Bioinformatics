# Running Mr. Bait

<br> **Citation:** *Chafin, Tyler and Douglas, Marlis and Douglas, Michael,2018. MrBait: Universal identification and design of targeted-enrichment capture probes.34,Bioinformatics. 10.1093/bioinformatics/bty548*
<br> **Description:** A software pipeline for identifying regions of interest in DNA sequence data and designing probes to enrich them.

## The workflow is divided into 5 steps, as follows:

1. Alignments (provided as .xmfa, .loci, or .maf) or genomes (provided as .fasta, annotated with .vcf or .gff) will be used to build a consensus sequence of each locus.
2. A sliding window will be applied to each consensus to find candidate targets for which baits could be designed
3. Targets are then selected (if too close together, or only one allowed per locus), and filtered according to any number of specified filter (e.g. GC content, flanking SNPs, pairwise alignment)
4. Passing targets are then parsed to design a putative set of baits
5. Baits are then filtered according to selected criteria, and output as FASTA.
6. The pipeline can be resumed and any steps iteratively re-visited by providing the SQLite database file (resulting in a significant reduction in runtime for successive runs)

### Step 1: loading alignments

To generate baits of length 80, tiled across target regions with an overlap of 40 bases, from a Multiple Alignment File (MAF) “example.maf”:
<br>`mrbait -M example.maf -b 80 -s tile=40`

<br>Bait selection scheme: Default behavior is to tile baits across all targets with 50% overlap
<br>
`mrbait -A ../data/references/Cacao_genomes/Theobroma_cacao_criollo_scaffold.v2.0.fna -T 6 -b 80 -t --strand "both" -o scaffTcacao `

- `--strand`: Output strand: Use this if you want to print baits as-designed, or as
reverse complement. Possible values: “+” [default], “-” (reverse-complement),
or “both”.
